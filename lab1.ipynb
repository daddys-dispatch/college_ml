Develop a program to create histograms for all numerical features and analyze the distribution of each feature. Generate box plots for all numerical features and identify any outliers. Use California Housing dataset.

```python
# Import required packages and dataset
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb

from sklearn.datasets import fetch_california_housing
```

```python
# Load the California Housing dataset
data = fetch_california_housing(as_frame=True)
housing_df = data.frame
```

```python
# Create histograms for numerical features
numerical_features = housing_df.select_dtypes(include=[np.number]).columns
```

```python
# Plot histograms
plt.figure(figsize=(15, 10))

for i, feature in enumerate(numerical_features):
    plt.subplot(3, 3, i + 1)
    sb.histplot(housing_df[feature], kde=True, bins=30)
    plt.title(f'Distribution of {feature}')
    
plt.tight_layout()
plt.show()
```

```python
# Generate box plots for numerical features
plt.figure(figsize=(15, 10))

for i, feature in enumerate(numerical_features):
    plt.subplot(3, 3, i + 1)
    sb.boxplot(x=housing_df[feature])
    plt.title(f"Box plot of {feature}")

plt.tight_layout()
plt.show()
```

```python
# Identify outliers using the IQR method
print("Outlier Detection:\n")
outliers_summary = {}

for feature in numerical_features:
    q1 = housing_df[feature].quantile(0.25)
    q3 = housing_df[feature].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    outliers = housing_df[(housing_df[feature] < lower_bound) | (housing_df[feature] > upper_bound)]
    outliers_summary[feature] = len(outliers)
    print(f"{feature}: {len(outliers)} outliers")
```

```python
# Print a summary of the dataset
print("Dataset Summary:\n")
print(housing_df.describe())
```
