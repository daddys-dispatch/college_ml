Develop a program to compute the correlation matrix to understand the relationships between pairs of features. Visualize the correlation matrix using a heatmap to know which variables have strong positive or negative correlations. Create a pair plot to visualize pairwise relationships between features. Use California Housing dataset.

```python
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import fetch_california_housing
```

```python
housing_df = fetch_california_housing(as_frame=True).frame
corr_matrix = housing_df.corr()
```

```python
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of California Housing Features")
plt.show()
```

```python
sns.pairplot(housing_df, diag_kind="kde")
plt.suptitle("Pair Plot of California Housing Features", y=1)
plt.show()
```
