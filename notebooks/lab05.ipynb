Develop a program to implement k-Nearest Neighbour (kNN) algorithm to classify the randomly generated 100 values of x in the range of [0,1]. Perform the following based on dataset generated:

- Label the first 50 points, x<sub>1</sub>...x<sub>50</sub>, as follows: If x<sub>i</sub> ≤ 0.5, then x<sub>i</sub> ∊ Class 1 else x<sub>i</sub> ∊ Class 2
- Classify the remaining points, x<sub>51</sub>...x<sub>100</sub>, using kNN for k = {1, 2, 3, 4, 5, 20, 30}

```python
import numpy as np
from collections import Counter
```

```python
data = np.random.rand(100)
train_data, test_data = data[:50], data[50:]
train_labels = ["Class 1" if x <= 0.5 else "Class 2" for x in train_data]
```

```python
def knn(train_data, train_labels, point, k):
    i = np.argsort(np.abs(train_data - point))[:k]
    return Counter(np.array(train_labels)[i]).most_common(1)[0][0]
```

```python
k_values = [1, 2, 3, 4, 5, 20, 30]

for k in k_values:
    print(f"\nk = {k}")
    for i, x in enumerate(test_data, start=51):
        print(f"x{i} ({x:.4f}) → {knn(train_data, train_labels, x, k)}")
```
